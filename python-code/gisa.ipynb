```python
data = []
for line in lines:
    aa = []
    temp = line[:-1].split(',')
    # print(temp)
    aa.append(int(temp[0]))
    aa.append(temp[1])
    for ele in temp[2:-3]:
        aa.append(int(ele))
    for i in temp[-3:]:
        aa.append(i)
    # print(aa)
    data.append(aa)
```

```python
# 1. # 파일에서 모든 라인을 가져온다.
with open('.\data\Abc1115.csv', 'r') as file:
    # 1-1. 파일 안의 컨텐츠를 연결하여 모든 라인을 가져온다.
    lines = file.readlines()

# 위 의 lines를 2차원 리스트로 만드시오
# 먼저 해야 할 일을 차례로 적는다.
# 해당 차례로 파이썬 코드로 옮긴다.
# 해야 할 일은 의미있는 작업을 하는 최소한(기본적)의 기능을 선택한다.

# 2. 한 줄을 가져온다.
# line = lines[0]
students = []
for line in lines:
    student = []
    
    # 3. 가져온 한 줄을 분리한다.(콤마로 분리)
    temp = line[:-1].split(',')
    
    # 4. 분리된 데이터를 타입에 맞추어 분리한다.
    stdNo = int(temp[0])
    student.append(stdNo)
    student.append(temp[1])
    for ele in temp[2:8]:
        val = int(ele)
        student.append(val)
        
    # 5. 분리한 데이터를 리스트에 저장한다.
    for ele in temp[8:]:
        student.append(ele)
        
    # 6. 리스트를 리스트에 저장한다.
    students.append(student)
    
# 7. 가지고 있는 줄(1000) 만큼 반복한다.
print(len(students))
```
